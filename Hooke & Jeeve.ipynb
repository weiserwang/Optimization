```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    
    return 3*x[0]**2-2*x[0]*x[1]+x[1]**2+4*x[0]+3*x[1]
```

```python
delta, epsilon, xk = 1, 0.5, np.array([0, 0])
def Hooke_Jeeve(delta,epsilon,xk):
    """
    Parameters
    ----------
    delta : The initial step length
    epsilon : The stopping criterion
    xk : The initial base point
    
    """

    yk = xk.copy()

    # The dim of the inital point
    dim = len(xk)

    # Initialize the number of iteration
    k = 1

    while delta >= epsilon:
        # Print all the information about this iteration
        print('The', k, 'th iteration')
        print('Base point:', xk)
        print('Funciton value at base point:', function(xk))
        print('Exploratory point:', yk)
        print('Funciton value at exploratory point', function(yk))
        print('Step length:', delta)

        # Exploratory move
        for i in range(dim):
            e = np.zeros([1, dim])[0]
            e[i] = 1

            # Compute and compare the function value 
            t1, t2 = function(yk + delta * e), function(yk)
            if t1 < t2:
                yk = yk + delta * e
            else:
                t1, t2 = function(yk - delta * e), function(yk)
                if t1 < t2:
                    yk = yk - delta * e
            print('The', i + 1, 'th point is', yk)
            print('The function value at that point:', function(yk))

        # Compute the new base point and exploratory point
        t1, t2 = function(yk), function(xk)
        if t1 < t2:
            xk, yk = yk, 2*yk - xk
        else:
            delta, yk = delta * 0.5, xk
        k += 1

        print("\n")
```

```python
Hooke_Jeeve(delta,epsilon,xk)
```

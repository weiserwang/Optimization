```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    a = x[0]
    b = x[1]
    return a**4+2*a**2*b**2+b**4
```

```python
def gradient_func(x):
    a = x[0]
    b = x[1]
    return np.array([4*a**3+4*a*b**2,4*a**2*b+4*b**3],dtype=float)
```

```python
def hessian_func(x):
    a = x[0]
    b = x[1]
    return np.array([[12*a**2+4*b**2,8*a*b],
                     [8*a*b,4*a**2+12*b**2]],dtype=float)
```

# Gradient Descent

```python
def GD(x,iterations,stepsize):
    k = 0
    num_of_iterations = iterations
    a = stepsize
    f_values = [f(x)]

    while k<num_of_iterations:
        d = -a*gradient_func(x)
        x += d
    #     print('X after iteration: {}'.format(x))
    #     print('Gradient value:{}'.format(gradient_func(x)))
        f_values.append(f(x))
        k += 1
        
    return x,f_values
```

```python
x_gd = np.array([1.,1.])
iterations = 30
stepsize = 0.01
x_gd,f_values = GD(x_gd,iterations,stepsize)

plt.plot(np.arange(iterations+1),f_values,marker='x',label='Newton Method')
plt.xlabel('Number of Iterations')
plt.ylabel('Function Value')
plt.title('Gradient Descent Method')
plt.legend()
```

# Newton Method

```python
x_newton = np.array([1.,1.])
iterations = 30

def newton(x,iterations):
    k = 0
    num_of_iterations = iterations
    f_values = [f(x)]

    while k<num_of_iterations:
        d = -1*np.dot(np.linalg.inv(hessian_func(x)),gradient_func(x))
        x += d
    #     print('X after iteration: {}'.format(x))
    #     print('Gradient value:{}'.format(gradient_func(x)))
        f_values.append(f(x))

        k += 1
    return x,f_values

x_newton,f_values = newton(x_newton,30)
plt.plot(np.arange(iterations+1),f_values,marker='x',label='Newton Method')
plt.xlabel('Number of Iterations')
plt.ylabel('Function Value')
plt.title('Newton Method')
plt.legend()
```
